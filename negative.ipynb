```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


def transform(img):
    for i in range(0, img.shape[0]):
        for j in range(0, img.shape[1]):
            img[i][j] = 255-img[i][j]


img = cv2.imread("C:/Users/Beyza/Desktop/elma.jpeg", 0)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
plt.show()

transform(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
plt.show()
```
